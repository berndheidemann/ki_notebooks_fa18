# Forward Propagation zu Fuß

```
x = [8, 3]
```

```
w1 = [
      [3,-2,-1,3],
      [9,2,2,1]
]
```

```
w2 = [1,-1,3,0]
```

```
y = 78
```

```
a2=[]
for col in range(len(w1[0])):
  sum=0
  for row in range(len(w1)):
    sum=sum+w1[row][col]*x[row]
  if sum<0:
    a2.append(0)
  else:
    a2.append(sum)
a2
```

```
y_hat=0
for i in range(len(w2)):
  y_hat=y_hat+w2[i]*a2[i]
y_hat
```

# Mit Matrixmultiplikation

```
import numpy as np
```

```
x = np.array([8, 3])
```

```
w1 = np.array([
      [3,-2,-1,3],
      [9,2,2,1]
])
```

```
w2 = np.array([1,-1,3,0])
```

```
y = 78
```

```
z2=x@w1
```

```
def relu(x):
  return np.maximum(0,x)
```

```
a2=relu(z2)
```

```
z3=a2@w2
```

```
z3
```

```
y_hat=relu(z3)
```

```
y_hat
```

## als Einzeiler

```
y_hat=relu(relu(x@w1)@w2)
```

```
y_hat
```

# Fehler

```
loss=(y-relu(relu(x@w1)@w2))**2
```

# For- und Backpropagation mit Pytorch

```
import torch
```

```
x = torch.tensor([8.0, 3.0])
```

```
w1 = torch.tensor([
      [3.0,-2.0,-1.0,3.0],
      [9.0,2.0,2.0,1.0]
], requires_grad=True)
```

```
w2 = torch.tensor([1.0,-1.0,3.0,0.0], requires_grad=True)
```

```
def relu(x):
  return torch.max(torch.tensor(0.), x)
```

```
y_hat=relu(relu(x@w1)@w2)
```

```
y_hat
```

```
loss=(y-y_hat)**2
```

```
loss
```

```
loss.backward()
```

```
w1.grad
```

```
w2.grad
```
